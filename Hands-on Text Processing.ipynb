# 2. Preprocessing

```python
with open("teks_berita.txt", "r") as text_file:
    teks = text_file.read()
teks
```

```python
# split into sentences
from nltk import sent_tokenize
kalimat = sent_tokenize(teks)
```

```python
kalimat
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
clean_teks = [tokenizer.tokenize(split_kalimat) for split_kalimat in kalimat]
clean_teks
```

```python
kata = [[split_word.lower() for split_word in kalimat] for kalimat in clean_teks]
kata
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
stopwords.words('indonesian')
```

```python
list_stopword = ["virus"]
```

```python
filtered_words = [[split_word for split_word in kalimat if split_word not in list_stopword] for kalimat in kata]
filtered_words
```

```python
filtered_words = [[split_word for split_word in kalimat if split_word not in stopwords.words('english')] for kalimat in kata]
```

```python
filtered_words
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lm = WordNetLemmatizer()
```

```python
lemmatized = [[lm.lemmatize(split_word) for split_word in words] for words in filtered_words]
lemmatized
```

```python
import pandas as pd
df = pd.DataFrame(columns=["asli","tanpa_stopword","lemmatized"])
df
```

```python
df['asli'] = pd.Series(kata)
df['tanpa_stopword'] = pd.Series(filtered_words)
df['lemmatized'] = pd.Series(lemmatized)
df
```

```python
df.to_csv("teks_berita.csv")
```
