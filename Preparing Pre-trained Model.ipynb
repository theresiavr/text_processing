#### Load libraries we will need 

```python
import pandas as pd
```

```python
df_news = pd.read_json("News_Category_Dataset_v2.json", lines = True)
df_news.columns
```

```python
df_news.head()
```

#### Data pre-processing

```python
from nltk import sent_tokenize
df_news["short_description_tokenized"] = df_news.short_description.apply(sent_tokenize)
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
df_news["short_description_tokenized"] = df_news.short_description_tokenized.apply(lambda x: [tokenizer.tokenize(split_kalimat) for split_kalimat in x])
df_news
```

```python
df_news["short_description_tokenized"] = df_news.short_description_tokenized.apply(lambda x: [[split_word.lower() for split_word in kalimat] for kalimat in x])
df_news
```

```python
from nltk.corpus import stopwords
```

```python
df_news["text"] = df_news.short_description_tokenized.apply(lambda x: [[split_word for split_word in kalimat if split_word not in stopwords.words('english')] for kalimat in x])
df_news
```

```python
from nltk.stem import WordNetLemmatizer
lm = WordNetLemmatizer()
```

```python
df_news["text"] = df_news.text.apply(lambda x: [[lm.lemmatize(split_word) for split_word in words] for words in x])
df_news
```

```python
df_news["joined_text"] = df_news["text"].apply(lambda x: [" ".join(sent) for sent in x])
df_news
```

```python
df_news["joined_text"] = df_news["joined_text"].apply(" ".join)
df_news
```

#### Processing the Data and TF-IDF

We need to merge the categories _WORDLPOST_ with _THE WORDPOST_, because there *are* basically the same, next we combine the columns _headline with _short_description_ these are our predictor text

```python
df_news.category = df_news.category.map(lambda x: "WORLDPOST" if x == "THE WORLDPOST" else x)
```

```python
df_news.to_csv("preprocessed_train.csv")
```

# ML

```python
from sklearn import model_selection
Train_X, Test_X, Train_Y, Test_Y = model_selection.train_test_split(df_news['joined_text'],df_news['category'],test_size=0.1)
```

```python
from sklearn.preprocessing import LabelEncoder
Encoder = LabelEncoder()
Train_Y = Encoder.fit_transform(Train_Y)
Test_Y = Encoder.fit_transform(Test_Y)
```

```python
filename = "encoder.sav"
with open(filename, 'wb') as handle:
    pickle.dump(Encoder, handle)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
Tfidf_vect = TfidfVectorizer(max_features=5000)
Tfidf_vect.fit(df_news['joined_text'])
Train_X_Tfidf = Tfidf_vect.transform(Train_X)
Test_X_Tfidf = Tfidf_vect.transform(Test_X)
```

```python
import pickle
filename = "vectorizer.sav"
with open(filename, 'wb') as handle:
    pickle.dump(Tfidf_vect, handle)
```

```python
print(Tfidf_vect.vocabulary_)
```

```python
print(Train_X_Tfidf)
```

```python
from sklearn import naive_bayes, svm
from sklearn.metrics import accuracy_score
# fit the training dataset on the NB classifier
Naive = naive_bayes.MultinomialNB()
Naive.fit(Train_X_Tfidf,Train_Y)
# predict the labels on validation dataset
predictions_NB = Naive.predict(Test_X_Tfidf)
# Use accuracy_score function to get the accuracy
print("Naive Bayes Accuracy Score -> ",accuracy_score(predictions_NB, Test_Y)*100)
```

```python
# Classifier - Algorithm - SVM
# fit the training dataset on the classifier
SVM = svm.LinearSVC(C=1.0, class_weight='balanced', multi_class='ovr', random_state=40, max_iter=10000)
SVM.fit(Train_X_Tfidf,Train_Y)
# predict the labels on validation dataset
predictions_SVM = SVM.predict(Test_X_Tfidf)
# Use accuracy_score function to get the accuracy
print("SVM Accuracy Score -> ",accuracy_score(predictions_SVM, Test_Y)*100)
```
